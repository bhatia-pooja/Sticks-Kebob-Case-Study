```python
file_path = 'Copy of stickskebob.xlsx'
xls = pd.ExcelFile(file_path)
```

```python
customer_survey_df = pd.read_excel(xls, 'Customer Survey')
```

```python
# Identify columns related to profession
profession_columns = [col for col in customer_survey_df.columns if 'What is your profession?' in col]

# Create a single 'Profession' column by filling in the selected choice
customer_survey_df['Profession'] = customer_survey_df[profession_columns].apply(lambda row: row.dropna().index[0] if row.notna().any() else None, axis=1)

# Drop the original profession columns
customer_survey_df.drop(columns=profession_columns, inplace=True)

# Display the dataframe to check the new 'Profession' column
customer_survey_df.head()
```

```python
profession_mapping = {
    'What is your profession? - Business professional': 7,
    'What is your profession? - Homemaker': 6,
    'What is your profession? - Design': 4,
    'What is your profession? - Educator': 5,
    'What is your profession? - General purpose contractor': 3,
    'What is your profession? - Accountant': 1,
    'What is your profession? - Lawyer': 2,
}

# Apply the updated mapping to the 'Profession' column
customer_survey_df['Profession'] = customer_survey_df['Profession'].map(profession_mapping)

# Display the updated dataframe to verify changes
customer_survey_df.head()
```
